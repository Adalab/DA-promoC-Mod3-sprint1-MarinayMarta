Vuestro set de datos debería tener al menos una variable categórica, el objetivo del pair programming de hoy:

```python
# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd

# Para la codificación de las variables numéricas
# -----------------------------------------------------------------------
from sklearn.preprocessing import LabelEncoder # para realizar el Label Encoding 
from sklearn.preprocessing import OneHotEncoder  # para realizar el One-Hot Encoding
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder  

# Para evitar que salgan los warnings en jupyter
# -----------------------------------------------------------------------
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/mnt/d/Adalab/ejercicios-de-adalab/modulo-3/DA-promoC-Mod3-sprint1-MarinayMarta/files/wellness_5.csv', index_col = 0)
df.head(2)
```

```python
# cambiamos el tipo de datos a 3 columnas:
for col in['DAILY_STRESS', 'BMI_RANGE', 'SUFFICIENT_INCOME']:
    df[col] = df[col].astype('category') 
```

```python
df['DAILY_STRESS'].replace(['1/1/00'], '3', inplace = True)
```

```python
df1= df.copy()
```

```python
df1.info()
```

```python
for col in df1.select_dtypes(exclude= 'number'):
  print(f'{[col]} : {df1[col].unique()}')
```

--> Las columnas que necesitaríamos codificar serian 'AGE' (c/ orden) y 'GENDER'(s/ orden)

* 'AGE': 
- => Label-Encoding

```python
# iniciamos el método
le = LabelEncoder()

df1['AGE_le']=le.fit_transform(df1['AGE'])
```

```python
df1.sample(4)
```

- --> nos ha codificado sin el orden esperado: la posición 0 deberia ser age < 20.
- --> lo intentamos con map

* 'AGE':
- => map

```python
#Lo primero, chequeamos cuales son los valores únicos.
df1.AGE.unique()
```

```python
# creamos un diccionario con el que mapearemos nuestra columna

mapa = {'Less than 20':0, '21 to 35':1, '36 to 50':2, '51 or more':3} 
```

```python
# aplicamos el map a nuestra columna, sobrescribimos:

df1["AGE"] = df1["AGE"].map(mapa)
# df1["AGE_map"] = df1["AGE"].map(mapa)
```

```python
df1.sample(6)
```

--> ha salido bien codificada.

* 'GENDER':
- => OneHotEncoding   (//out)

```python
# iniciamos el método de OneHot Encoder
oh = OneHotEncoder()
```

```python
# en este caso trabajaremos con la columna english, que en este caso no tiene orden

df1.GENDER.unique()
```

```python
# hacemos la codificación de los datos para la variable dada 
transformados = oh.fit_transform(df1[['GENDER']])
```

```python
# convertimos nuestro array con la codificación hecha en un dataframe, donde tendremos dos columnas, una para los "Female" y otra para los "Male"
oh_df1 = pd.DataFrame(transformados.toarray())
oh_df1.head()
```

```python
# el método get_feature_names nos va a dar el nombre de las columnas nuevas que se nos generarán
oh_df1.columns = oh.get_feature_names()
oh_df1.columns
```

```python
# concatenamos el dataframe original con el dataframe que acabamos de crear
final = pd.concat([df1, oh_df1],axis=1)
```

```python
final.head()
```

* 'GENDER'
- --> get_dummies

```python
#
dummies = pd.get_dummies(df1['GENDER'], prefix_sep = "_", prefix = "owners", dtype = int)
dummies.head(2)
```

```python
# juntamos los datos condf_dummies
df_dummies = pd.concat([df1, dummies], axis = 1)
df_dummies.head(2)
```
